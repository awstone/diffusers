```python
from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained("./output", torch_dtype=torch.float16, use_safetensors=True).to("cuda")
image = pipeline("A photo of sks dog in a bucket", num_inference_steps=50, guidance_scale=7.5).images[0]
image
```

```python
from diffusers import FluxPipeline

pipeline = FluxPipeline.from_pretrained('black-forest-labs/FLUX.1-dev', torch_dtype=torch.float16, use_safetensors=True).to("cuda")
pipeline.load_lora_weights("flux-dreambooth-lora", weight_name="pytorch_lora_weights.safetensors")
image = pipeline("A photo of sks dog in downtown manhattan").images[0]
image
```
